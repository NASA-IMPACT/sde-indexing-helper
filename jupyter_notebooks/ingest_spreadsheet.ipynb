```python
def combine_notes(row):
    return "Notes From Test Server QC:\n" + str(row["Notes From Test Server QC"]) + "\n\n" + "Consolidated Notes:\n" + str(row["Consolidated Notes"]) + "\n\n" + "Notes:\n" + str(row["Notes"]) + "\n\n" + "Updated Notes:\n" + str(row["Updated Notes"]) + "\n\n" + "Shravan Notes:\n" + str(row["Shravan Notes"]) + "\n\n"

```

```python
df['final_notes'] = df.apply(lambda row: combine_notes(row), axis=1)
```
